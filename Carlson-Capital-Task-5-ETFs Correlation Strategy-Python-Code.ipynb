```python
data1['change_previous_day'] =0
data1['change_pre_market'] =0
for i in range(1,len(data1)):
  data1['change_previous_day'].iloc[i]=(((data1['Open'].iloc[i]-data1['Close'].iloc[i-1])/data1['Close'].iloc[i-1])*100)
for i in range(len(data1)):
  data1['change_pre_market'].iloc[i]=(((data1['Close'].iloc[i]-data1['Open'].iloc[i])/data1['Open'].iloc[i])*100)
```

## Back Testing Model (with Dynamic Correlation feature)

```python
capital=1000
returns_Long=[] #List storing returns from long trades
returns_Short=[] #List storing returns from short trades
long_trades=0
short_trades=0
etf=[]
etf2=[]
k=0
position=0
stop_loss_percent=0.1
twd1=pd.DataFrame(columns=['ETF','Entry Date','Entry Price','Exit Price','Returns for Trade in %','Type of Trade','Trade Close By'])
for i in range(3,len(data1)):

  a=(i)*(6.5)*(60/interval_etf)

  df_etfs=df_final2.copy()
  df_sp500=sp500_data.copy()

  df_etfs=df_etfs.drop(df_etfs.index[i+1:len(df_final2)])
  df_sp500=df_sp500.drop(df_sp500.index[i+1:len(df_final2)])

  correlations=df_etfs.corrwith(df_sp500)
  highest_corr_etf1=correlations.idxmax()
  etf2.append(highest_corr_etf1)

  data_etf=dict1[highest_corr_etf1]
  # data_etf.reset_index(inplace=True)
  # data_etf['Date'] = df['Datetime'].dt.date
  # df['Time'] = df['Datetime'].dt.time
  data_f=data1.copy()

  if(data_f['change_pre_market'].iloc[i]>=0.5):
    new_row={'ETF':highest_corr_etf1,'Entry Date':data_f['Date'].iloc[i],'Entry Price':data_etf['Open'].iloc[a],'Exit Price':0,'Returns for Trade in %':0,'Type of Trade':"long",'Trade Close By':"none"}
    twd1.loc[len(twd1)]=new_row
    position=1
    # twd1['ETF'].iloc[k]=highest_corr_etf1
    # twd1['Entry Date'].iloc[k]=data_f['Date'].iloc[i]
    # twd1['Entry Price'].iloc[k]=data_etf['Open'].iloc[a]
    # twd1['Type of Trade'].iloc[k]='long'
    slp=data_etf['Open'].iloc[a]*(1-stop_loss_percent)
    for j in range(a+1,a+(30/interval_etf)+1):
      if(data_etf['Close'].iloc[j]<=slp and position==1): #Close the long trade by stop loss condition
        twd1['Exit Price'].iloc[k]=data_etf['Close'].iloc[j]
        twd1['Trade Close By'].iloc[k]='stop_loss'
        twd1['Returns for Trade in %']=(((data_etf['Close'].iloc[j]-data_etf['Open'].iloc[a])/data_etf['Open'].iloc[a])*100)
        returns_Long.append(((data_etf['Close'].iloc[a+(30/interval_etf)]-data_etf['Open'].iloc[a])/data_etf['Open'].iloc[a])*100)
        position=0
        k+=1
        break
    if(position==1): #Square off the long trade after 30 minutes if not closed by stop loss
      twd1['Exit Price'].iloc[k]=data_etf['Close'].iloc[a+(30/interval_etf)]
      twd1['Trade Close By'].iloc[k]='exit'
      twd1['Returns for Trade in %']=(((data_etf['Close'].iloc[a+(30/interval_etf)]-data_etf['Open'].iloc[a])/data_etf['Open'].iloc[a])*100)
      returns_Long.append(((data_etf['Close'].iloc[a+(30/interval_etf)]-data_etf['Open'].iloc[a])/data_etf['Open'].iloc[a])*100)
      position=0
    long_trades+=1
    etf.append(highest_corr_etf1)


  elif(data_f['change_pre_market'].iloc[i]<=-0.5):
    new_row={'ETF':highest_corr_etf1,'Entry Date':data_f['Date'].iloc[i],'Entry Price':data_etf['Open'].iloc[a],'Exit Price':0,'Returns for Trade in %':0,'Type of Trade':"short",'Trade Close By':"none"}
    twd1.loc[len(twd1)]=new_row
    position=-1
    # twd1['ETF'].iloc[k]=highest_corr_etf1
    # twd1['Entry Date'].iloc[k]=data_f['Date'].iloc[i]
    # twd1['Entry Price'].iloc[k]=data_etf['Open'].iloc[a]
    # twd1['Type of Trade'].iloc[k]='short'
    slp=data_etf['Open'].iloc[a]*(1+stop_loss_percent)
    for j in range(a+1,a+(30/interval_etf)+1):
      if(data_etf['Close'].iloc[j]>=slp and position==-1): #Close the short trade by stop loss condition
        twd1['Exit Price'].iloc[k]=data_etf['Close'].iloc[j]
        twd1['Trade Close By'].iloc[k]='stop_loss'
        twd1['Returns for Trade in %']=(((data_etf['Open'].iloc[a]-data_etf['Close'].iloc[j])/data_etf['Open'].iloc[a])*100)
        returns_Short.append(((data_etf['Open'].iloc[a]-data_etf['Close'].iloc[j])/data_etf['Open'].iloc[a])*100)
        position=0
        k+=1
        break
    if(position==-1): #Square off the short trade after 30 minutes if not closed by stop loss
      twd1['Exit Price'].iloc[k]=data_etf['Close'].iloc[a+(30/interval_etf)]
      twd1['Trade Close By'].iloc[k]='exit'
      twd1['Returns for Trade in %']=(((data_etf['Open'].iloc[a]-data_etf['Close'].iloc[a+(30/interval_etf)])/data_etf['Open'].iloc[a])*100)
      returns_Short.append(((data_etf['Open'].iloc[a]-data_etf['Close'].iloc[j])/data_etf['Open'].iloc[a])*100)
      position=0
    short_trades+=1
    etf.append(highest_corr_etf1)


# Final Results
print("No. of Long Trades:",long_trades)
print("No. of Short Trades:",short_trades)
print("Returns from Long Trades:",sum(returns_Long),'%')
print("Returns from Short Trades:",sum(returns_Short),'%')
print("Total Returns:",sum(returns_Long)+sum(returns_Short),'%')
print("Final Capital:$",((sum(returns_Long)+sum(returns_Short))*capital)/100)
print("Traded ETF List:",etf)
print("Dynamic ETF List with highest correlation with S&P500:",etf)
```
